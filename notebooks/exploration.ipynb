# Audio Data Exploration
Explore signal processing techniques like FFT and Mel-spectrograms.

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from src.preprocessing import load_audio, get_mel_spectrogram, get_fft

# Sample path (assuming a local data folder)
sample_path = 'data/sample1.wav'
```

## 1. Waveform Visualization

```python
audio, sr = load_audio(sample_path)
plt.figure(figsize=(10, 4))
librosa.display.waveshow(audio, sr=sr)
plt.title('Waveform: ' + sample_path)
plt.show()
```

## 2. Fast Fourier Transform (FFT)

```python
fft_vals = get_fft(audio)
frequencies = np.fft.rfftfreq(len(audio), 1/sr)

plt.figure(figsize=(10, 4))
plt.plot(frequencies, fft_vals)
plt.title('FFT Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```

## 3. Mel-Spectrogram Extraction

```python
mel_spec = get_mel_spectrogram(audio, sr)

plt.figure(figsize=(10, 4))
librosa.display.specshow(mel_spec[0].numpy(), sr=sr, x_axis='time', y_axis='mel')
plt.title('Mel-Spectrogram')
plt.colorbar(format='%+2.0f dB')
plt.show()
```
